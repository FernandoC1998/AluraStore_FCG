### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



### **Tienda 1**




```python
tienda['Totales'] = tienda['Precio'] * tienda['Cantidad de cuotas']
tienda.head()
```

```python
total_tienda1 = tienda['Totales'].sum()
total_tienda1
```

### **Tienda 2**

```python
tienda2['Totales'] = tienda2['Precio'] * tienda2['Cantidad de cuotas']
tienda.head()
```

```python
total_tienda2 = tienda2['Totales'].sum()
total_tienda2
```

### **Tienda 3**

```python
tienda3['Totales'] = tienda3['Precio'] * tienda3['Cantidad de cuotas']
tienda.head()
```

```python
total_tienda3 = tienda3['Totales'].sum()
total_tienda3
```

### **Tienda 4**

```python
tienda4['Totales'] = tienda4['Precio'] * tienda4['Cantidad de cuotas']
tienda.head()
```

```python
total_tienda4 = tienda4['Totales'].sum()
total_tienda4
```

### **Comparación grafica**

```python
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
totales = [total_tienda1 , total_tienda2 , total_tienda3 , total_tienda4]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
barras = plt.bar(tiendas, totales)

plt.title('Ventas Totales por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ventas Totales')

# Mostrar totales encima de cada barra
for barra in barras:
    height = barra.get_height()
    plt.text(
        barra.get_x() + barra.get_width() / 2,
        height,
        f'{height:,.0f}',
        ha='center',
        va='bottom'
    )

plt.show()
```

# 2. Ventas por categoría

### **Tienda 1**

```python
ventas_por_categoria1 = tienda.groupby('Categoría del Producto')['Totales'].sum()
ventas_por_categoria1
```

```python
plt.figure(figsize=(12,6))
bars = plt.bar(ventas_por_categoria1.index, ventas_por_categoria1.values)

plt.title('Ventas por Categoría - Tienda 1')
plt.xlabel('Categoría')
plt.ylabel('Ventas Totales')

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f'{height:,.0f}',
        ha='center',
        va='bottom'
    )

plt.xticks(rotation=45)
plt.show()
```
